1. Por meio do dataset paises.csv, trace dois gráficos de linhas em um mesmo plano cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da América do Norte;

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('paises.csv', delimiter=';')

pattern = 'NORTHERN AMERICA'
df_norte = df[df['Region'].str.contains(pattern, case=False, na=False)]

plt.figure(figsize=(10, 6))
plt.plot(df_norte['Country'], df_norte['Deathrate'], marker='o', linestyle='-', color='red', label='Deathrate')
plt.plot(df_norte['Country'], df_norte['Birthrate'], marker='o', linestyle='--', color='blue', label='Birthrate')

plt.xlabel('Países da América do Norte')
plt.ylabel('Taxa (%)')
plt.title('Taxa de Mortalidade e Taxa de Natalidade dos países da América do Norte')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

2. Por meio do dataset space.csv, trace um gráfico em barras
mostrando quantas empresas espaciais diferentes os EUA e a CHINA
possuem;
Dica: não se esqueça de retirar os resultados repetidos

```python
df_space = pd.read_csv('space.csv', delimiter=';')

df_space['Country'] = df_space['Location'].apply(lambda loc: loc.split()[-1] if isinstance(loc, str) else loc)

df_usa = df_space[df_space['Country'].str.upper() == 'USA']
df_china = df_space[df_space['Country'].str.upper() == 'CHINA']

unique_usa_companies = df_usa['Company Name'].drop_duplicates()
unique_china_companies = df_china['Company Name'].drop_duplicates()

count_usa = len(unique_usa_companies)
count_china = len(unique_china_companies)

countries = ['USA', 'China']
counts = [count_usa, count_china]

plt.figure(figsize=(6, 4))
plt.bar(countries, counts, color=['blue', 'red'])
plt.xlabel('País')
plt.ylabel('Número de Empresas Espaciais Únicas')
plt.title('Empresas Espaciais Únicas - USA vs China')
plt.tight_layout()
plt.show()
```

3. Por meio do dataset space.csv, trace um gráfico em torta ilustrando a porcentagem de missões da empresa Roscosmos que deram certo e que deram errado;

```python
df_roscosmos = df_space[df_space['Company Name'].str.upper() == 'ROSCOSMOS']

outcome_counts = df_roscosmos['Status Mission'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(outcome_counts.values, labels=outcome_counts.index, autopct='%1.1f%%')
plt.title('Porcentagem de Missões da Roscosmos:\nSucesso vs Falha')
plt.show()
```
